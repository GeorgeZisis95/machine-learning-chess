```python
import torch
import matplotlib.pyplot as plt
```

```python
# Load the losses from .pt files
loss1 = torch.load('data/loss/loss_log_1.pt')
loss2 = torch.load('data/loss/loss_log_2.pt')
loss3 = torch.load('data/loss/loss_log_3.pt')
```

```python
loss1 = loss1["train_losses"]
loss2 = loss2["train_losses"]
loss3 = loss3["train_losses"]
```

```python
# Convert to numpy arrays if they are tensors
loss1 = loss1.numpy() if isinstance(loss1, torch.Tensor) else loss1
loss2 = loss2.numpy() if isinstance(loss2, torch.Tensor) else loss2
loss3 = loss3.numpy() if isinstance(loss3, torch.Tensor) else loss3

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(loss1, label='ConvNet with positional planes only')
plt.plot(loss2, label='ConvNet with the addition of helper planes')
plt.plot(loss3, label='ResNet')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Comparison')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
